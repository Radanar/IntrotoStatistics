```python
import math
import scipy.stats
import numpy
import matplotlib.pyplot as plt

group1 = [3.6, 2.9, 3.3, 2.5, 3.8]
group2 = [2.1, 2.6,	3.1, 3.9, 3.4]
group3 = [4.0, 2.0,	3.2, 2.6, 3.2]
group4 = [2.0, 3.6, 2.7, 3.9, 2.5]

n1 = len(group1)
n2 = len(group2)
n3 = len(group3)
n4 = len(group4)

n = n1+n2++n3+n4

ssBetween = (sum(group1)**2/n1)\
      +(sum(group2)**2/n2)\
      +(sum(group3)**2/n3)\
      +(sum(group4)**2/n4)\
      -((sum(group1)+sum(group2)+sum(group3)+sum(group4))**2/n)

x = numpy.linspace(0,4,100)

ssTotal = sum(numpy.array(group1)**2)\
      +sum(numpy.array(group2)**2)\
      +sum(numpy.array(group3)**2)\
      +sum(numpy.array(group4)**2)\
      -(sum(group1)+sum(group2)+sum(group3)+sum(group4))**2/n

ssWithin = ssTotal - ssBetween

k = 4

dfn = k - 1
dfd = n - k
msBetween = ssBetween/dfn
msWithin = ssWithin/dfd

f_ratio = msBetween/msWithin
print("F-factor is",f_ratio)

plt.plot(x, scipy.stats.f.pdf(x,dfn,dfd,loc=0,scale=1))
plt.show()

1-scipy.stats.f.cdf(f_ratio,dfn,dfd,loc=0,scale=1)

print(" Among the 20 samples, if we look at the measurements, the numbers around 20 to 23 are more than the numbers higher than 24 in the sample as below-\
2.0 2.0 2.1	2.5 2.5	2.6 2.6 2.7 , 2.9 3.1	3.2	3.2	3.3	3.4 3.6 3.6 3.8	3.9 3.9 4.0 \
So, we can see that there is no balance in the sample which is the numbers around 2 GPA are less than the number higer than  2.7 GPA \
and that shows that there is a significant difference in the sample.")

print("Null Hypothesis: The samples of GPA among the teams  does not have differences.")
print("We will reject the null hypothesis as f-ratio is more than the the alpha value 0.05. And, the sample averages are significantly different from each other")
```
