```python
import scipy.stats
import math

n1 = 10
n2 = 15

s1 = 1.6
s2 = 1.6

diff = 0.6

stdEr = math.sqrt(s1**2/n1+s2**2/n2)
print("Standard Error:", stdEr)

df = (s1**2/n1 + s2**2/n2)**2/((1/(n1-1))*(s1**2/n1)**2 + (1/(n2-1))*(s2**2/n2)**2)
print("The degree of freedom is:",df)

pvalue  = 1-scipy.stats.t.cdf(diff,df,loc = 0, scale = stdEr)

print("The probability > ",diff," is:", pvalue, "\n")
print("Hypothesis: There is difference in the graduation time. \n")
print("Null Hypothesis: There is no difference in the graduation time. \n")
print("For alpha 0.01, We will accept the null hypothesis as P-value", pvalue , "is more than alpha 0.01. To reject the H0, p-value should be less than 0.01\n")

print("At 5% level of significance, these 2 colleges are not different as p-value of",pvalue,"is greater than alpha value 0.05")
```
