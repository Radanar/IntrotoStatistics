```python
import math
import scipy.stats


ss_total =  621.4
ss_within = 374.5 

ss_between = ss_total - ss_within
print("ss_between =", ss_between)

k = 4
n = 30
ms_factor = ss_between/(k-1)
print("ms_factor =", ms_factor)

ms_error = ss_within/(n-k)
print("ms_error =", ms_error)

f_ratio = ms_factor/ms_error
print("f_ratio =", f_ratio)

print("We will reject the null hypothesis as f-ratio is far greater than the alpha value 0.01. The null hypothesis is false and all the samples have great difference between them.")
```
