```python
import math
import scipy.stats
import numpy
import matplotlib.pyplot as plt

k = 5
group1 = [2625,2997,4915]
group2 = [5348,5682,5482]
group3 = [6583,8560,3830]
group4 = [7285,6897,9230]
group5 = [6277,7818,8677]

n1 = len(group1)
n2 = len(group2)
n3 = len(group3)
n4 = len(group4)
n5 = len(group5)
n = n1+n2+n3+n4+n5

ssBetween = sum(group1)**2/n1\
      +sum(group2)**2/n2\
      +sum(group3)**2/n3\
      +sum(group4)**2/n4\
      +sum(group5)**2/n5\
      -((sum(group1)+sum(group2)+sum(group3)+sum(group4)+sum(group5))**2/n)

dfn = k - 1
msBetween = ssBetween/dfn

ssTotal = sum(numpy.array(group1)**2)\
      +sum(numpy.array(group2)**2)\
      +sum(numpy.array(group3)**2)\
      +sum(numpy.array(group4)**2)\
      +sum(numpy.array(group5)**2)\
      -(sum(group1)+sum(group2)+sum(group3)+sum(group4)+sum(group5))**2/n

ssWithin = ssTotal - ssBetween

dfd = n - k
msWithin = ssWithin/dfd

f_factor = msBetween/msWithin
print("Sayar's Version_Just Example")
print("The F factor is",f_factor)

x = numpy.linspace(0,5,100)

plt.plot(x, scipy.stats.f.pdf(x,dfn,dfd,loc=0,scale=1))
plt.show()

1-scipy.stats.f.cdf(f_factor,dfn,dfd,loc=0,scale=1)
```

```python
import math
import scipy.stats
import numpy
import matplotlib.pyplot as plt

group1 = [9, 66, 98]
group2 = [16, 93, 82]
group3 = [22, 147, 120]
group4 = [30, 199, 148]
group5 = [27, 168, 132]

n1 = len(group1)
n2 = len(group2)
n3 = len(group3)
n4 = len(group4)
n5 = len(group5)
n = n1+n2++n3+n4+n5

ssBetween = (sum(group1)**2/n1)\
      +(sum(group2)**2/n2)\
      +(sum(group3)**2/n3)\
      +(sum(group4)**2/n4)\
      +(sum(group5)**2/n5)\
      -((sum(group1)+sum(group2)+sum(group3)+sum(group4)+sum(group5))**2/n)

x = numpy.linspace(0,5,100)

ssTotal = sum(numpy.array(group1)**2)\
      +sum(numpy.array(group2)**2)\
      +sum(numpy.array(group3)**2)\
      +sum(numpy.array(group4)**2)\
      +sum(numpy.array(group5)**2)\
      -(sum(group1)+sum(group2)+sum(group3)+sum(group4)+sum(group5))**2/n

ssWithin = ssTotal - ssBetween

k = 5

dfn = k - 1
dfd = n - k
msBetween = ssBetween/dfn
msWithin = ssWithin/dfd

f_ratio = msBetween/msWithin
print("F-factor is",f_ratio)

plt.plot(x, scipy.stats.f.pdf(x,dfn,dfd,loc=0,scale=1))
plt.show()

1-scipy.stats.f.cdf(f_ratio,dfn,dfd,loc=0,scale=1)

print("Null Hypothesis: The colony in patients does not tested MRSA.")
print("We will accept the null hypothesis as f-ratio is more than the the alpha value 0.05 as the result is insignificant.")
```
