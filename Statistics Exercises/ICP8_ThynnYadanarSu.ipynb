```python
import scipy.stats
import math

n1 = 10
n2 = 15

s1 = 1.6
s2 = 1.6

diff = 0.6

stdEr = math.sqrt(s1**2/n1+s2**2/n2)
print("Standard Error:", stdEr)

df = (s1**2/n1 + s2**2/n2)**2/((1/(n1-1))*(s1**2/n1)**2 + (1/(n2-1))*(s2**2/n2)**2)
print("The degree of freedom is:",df)

pvalue  = 1-scipy.stats.t.cdf(diff,df,loc = 0, scale = stdEr)

print("The probability > ",diff," is:", pvalue, "\n")
print("Hypothesis: Wax 1 lasts longer than Wax 2. \n")
print("Null Hypothesis: Wax 1 doesn't last longer than Wax 2. \n")
print("We will accept the null hypothesis as P-value is more than alpha 0.02. \n")
```
