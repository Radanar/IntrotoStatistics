```python
#Sample
import scipy
from scipy.stats import t
import math

data = [ 10, 20, 30, 40] #create Python list
n = len(data)
print("No of data points:", n)

mean = sum(data)/n
print("Mean:", mean)

s = scipy.stats.tstd(data)
print("Standard deviation:", s)

lower_bound = scipy.stats.t.ppf(0.05, n-1, loc = mean, scale = s/math.sqrt(n))
upper_bound = scipy.stats.t.ppf(0.95, n-1, loc = mean, scale = s/math.sqrt(n))

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```
