```python
import math
import scipy.stats

s1=1
s2=1.2

n1=25
n2=16

diff =  0.3

stdEr = math.sqrt(s1**2/n1+s2**2/n2) 
print("Standard Error:", stdEr)

df = (s1**2/n1 + s2**2/n2)**2/((1/(n1-1))*(s1**2/n1)**2 + (1/(n2-1))*(s2**2/n2)**2)
print("Degree of Freedom: ", df)

percentage_less = scipy.stats.t.cdf(-diff,df,loc=0, scale= stdEr)
percentage_more = 1-scipy.stats.t.cdf(diff,df,loc=0, scale= stdEr)
print("Percentage less than", "-", diff, "=", percentage_less)
print("Percentage more than", diff, "=", percentage_more)

print("We will accept the null hypothesis as P-Value(",(percentage_less+percentage_more) ,") is larger than alpha 0.05.")
```
