```python
import math
import scipy.stats
import numpy
import matplotlib.pyplot as plt

group1 = [216,	198,	240,	187,	176]
group2 = [202,	213,	284,	228,	210]
group3 = [170,	165,	182,	197,	201]

n1 = len(group1)
n2 = len(group2)
n3 = len(group3)

n = n1+n2++n3

ssBetween = (sum(group1)**2/n1)\
      +(sum(group2)**2/n2)\
      +(sum(group3)**2/n3)\
      -((sum(group1)+sum(group2)+sum(group3))**2/n)

x = numpy.linspace(0,3,100)

ssTotal = sum(numpy.array(group1)**2)\
      +sum(numpy.array(group2)**2)\
      +sum(numpy.array(group3)**2)\
      -(sum(group1)+sum(group2)+sum(group3))**2/n

ssWithin = ssTotal - ssBetween

k = 3

dfn = k - 1
dfd = n - k
msBetween = ssBetween/dfn
msWithin = ssWithin/dfd

f_ratio = msBetween/msWithin
print("F-factor is",f_ratio)

plt.plot(x, scipy.stats.f.pdf(x,dfn,dfd,loc=0,scale=1))
plt.show()

1-scipy.stats.f.cdf(f_ratio,dfn,dfd,loc=0,scale=1)

print("Null Hypothesis: The three groups of mean does not have similar mean.")
print("F-ratio is greater than the alpha value 0.05, \
 the result is not significant and cannot reject the null hypothesis. ")
```
